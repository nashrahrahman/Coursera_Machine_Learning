```python
#computing cost J

def computeCost(x, y, theta):
    s = np.power(( x.dot(theta) - np.transpose([y]) ), 2)
    #print(s.shape)
   # print(type(s))
    J = (1.0/(2*m))*s.sum(axis=0)
    return J
```

```python
computeCost(x,y, theta)
```

```python
#### Gradient Descent
def gradientDescent(x,y,theta,alpha,iterations):
    
    m=len(y)
    J_history=np.zeros((iterations,1))
    
    for i in range(1,(iterations+1)):
        interm= ( x.dot(theta) - np.transpose([y]) )
        #print(x.shape)
        tranx=np.transpose(x)
        #print(tranx.shape)
        interm1= tranx.dot(interm) 
        theta = theta - (alpha*(1.0/m) * interm1)
        #print(theta.shape)
        #theta = theta - ((alpha/m)*(np.transpose(x).dot(x.dot(theta)-np.transpose(y))))
        #J_history[i]= float(computeCost(x,y, theta))
        
    
    return theta
```

```python
iterations = 1500
alpha = 0.01
theta = gradientDescent(x,y,theta,alpha,iterations)
print(theta.shape)
```
